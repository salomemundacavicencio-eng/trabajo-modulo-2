```python
import unittest
from gestor_contactos import GestorContactos
from contacto import Contacto

class TestGestorContactos(unittest.TestCase):

    def test_agregar_y_buscar(self):
        gestor = GestorContactos()
        contacto = Contacto("Ana", "123", "ana@mail.com", "Chile")
        gestor.agregar_contacto(contacto)
        self.assertIsNotNone(gestor.buscar_contacto("Ana"))

if __name__ == "__main__":
    unittest.main()
```
